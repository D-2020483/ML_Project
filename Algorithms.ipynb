```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
hours = np.array([1,2,3,4,5,6]).reshape(-1,1)
marks = np.array([25,45,55,65,75,85])
dataset = pd.DataFrame({
    'Hours': hours.flatten(),  
    'Marks': marks
})

print(dataset)
```

```python
#create a train model
model = LinearRegression()
model.fit(hours, marks)
```

```python
#make prediction for 07 th study hour
predicted = model.predict([[7]])
print("Predicted marks for 7 hours of study:", predicted[0])
print ("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

```python
plt.scatter(hours, marks, color='blue', label='Actual data')
plt.plot(hours, model.predict(hours), color='red', label='Regression line')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#create data set
hours = np.array([1,2,3,4,5,6]).reshape(-1, 1)
passed = np.array([0,0,0,1,1,1])
dataset = pd.DataFrame({
    'Hours': hours.flatten(),  
    'Passed': passed
})
print(dataset)
```

```python
#create and train model
model = LogisticRegression()
model.fit(hours, passed)
```

```python
#make prediction for student who stuied 4 hours
prediction = model.predict([[4]])
probability = model.predict_proba([[4]])
print("Predicted result (1=Pass, 0=Fail):", prediction[0])
print("Probability of passing:", probability[0][1])
```

```python
# Step 4: Visualize the results
plt.scatter(hours, passed, color='blue', label='Actual data')
plt.plot(hours, model.predict_proba(hours)[:, 1], color='red', label='Sigmoid curve')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression Example')
plt.legend()
plt.show()
```
